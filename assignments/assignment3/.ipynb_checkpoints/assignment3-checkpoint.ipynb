### Task 1: Text classification with Ollama (2p)

The `data/emails.csv` file contains 12 email headlines, with 4 spam emails, 4 legitimate work emails and 4 vague emails that are hard to classify based on the title alone. Use this dataset for all subtasks in this task.

a) Make a function for classifying emails (based on the headlines) as spam, work or unknown. The function should return only the classification and nothing else. (0.5p)

b) Use the smaller gemma3 (270m) to classify the emails using the function created in part a. (0.5p)

c) Use larger gemma3 (4b) to classify the emails using the function created in part a). In separate markdown cell, write a brief comment comparing the results of parts b) and c). (0.5p)

d) Write a script that repeats b) and c) 3 times, storing the results for both models separately. For both models, put the results as columns into a new DataFrame that also contains the headlines so that it is easy to compare how the output varied across runs for both models. Comment if there were differences and explain why this happened. (0.5p)

````python
import pandas as pd
emails = pd.read_csv("data/emails.csv", sep ="\t", header=0, names=["headline"],engine='python',encoding="utf-8") #to force to read one line as one email

def ask_ai(prompt):
    response = ollama_client.chat.completions.create(
        model=MODEL,
        messages=[{"role": "user", "content": prompt}])
    return response.choices[0].message.content.strip() #.strip() to avoid the \n at the end of each classification

#a
def classification (df):
    prompt = f"""Classify the emails of the following email dataframe:{df}
                Respond with exactly one words: Spam, Work, Unknown"""
    classification_response = ask_ai(prompt)
    return classification_response.strip()

#b
MODEL = "gemma3:270m"
emails["classification"]= emails["headline"].apply(classification)
display(emails)
````

````python
#c
MODEL = "gemma3:1b" #4b is too hard for my computer, I used only 1b
emails["classification"]= emails["headline"].apply(classification)
display(emails)
````

````python
#Result c seems much more accurate than b.
````

````python
#d
emails_b =pd.DataFrame(emails["headline"])
emails_c=pd.DataFrame(emails["headline"])

for repetition in range (1,4):
    MODEL = "gemma3:270m"
    emails_b[f"classification_{repetition}"]= emails_b["headline"].apply(classification)
    MODEL = "gemma3:1b"
    emails_c[f"classification_{repetition}"]= emails_c["headline"].apply(classification)

display("classification with gemma3 (270m):")
display (emails_b)
display ("classification with gemma3 (1b):")
display (emails_c)
````

### Task 2: Sentiment analysis with Ollama (2p)

The `data/news.csv` file contains 10 fictional financial news headlines. Use it for all subtasks in this task.

a) Make a function for classifying the texts in the provided dataset based on the topic (earnings, mergers, regulation, macroeconomics) and for determining the sentiment of the news (positive, negative, neutral). The function should return the class and sentiment in JSON format. (1p)

b) Use gemma3 (4b) to classify and provide the sentiment for each row of the provided dataset, inserting them into a new DataFrame that contains both the original headlines as well as topic and sentiment. (0.5p)

c) Give the same data and prompt to a browser based LLM (e.g. ChatGPT, LeChat, Claude or Gemini) and ask it to provide the topic and sentiment, giving it the same options. Paste the results into a markdown cell. Compare the results of b) and c), which one is more accurate and why? (0.5p)

````python
#a
import json

news = pd.read_csv("data/news.csv", encoding="utf-8", sep ="\t" )

def analyse_topic_sentiment(df):
    prompt=f"""Analyse the topic and the sentimet of the following news: {df}
    Return your answer as a JSON object with two keys:
    - "Topic": one of "Earning", "Mergers", "Regulation", or "Macroeconomics"
    - "Sentiment": one of "Positive", "Neutral", or "Negative"
    Do not include any other text, mardown formatting, or code blocks.
    Return only the JSON object."""
    result = ask_ai(prompt)
    cleaned = result.strip().removeprefix("```json").removeprefix("```").removesuffix("```").strip() #to avoid any problem
    return json.loads(cleaned)
    
````

````python
#b
MODEL = "gemma3:1b"

analyse = news["headline"].apply(analyse_topic_sentiment)

news["Topic"] = analyse.apply(lambda d: d.get("Topic", ""))
news["Sentiment"] =analyse.apply(lambda d: d.get("Sentiment", ""))

display (news)
````

````python
#a
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

bank_df = pd.read_csv("data/bank-additional/bank-additional.csv",sep=";", encoding = "utf-8")

display (bank_df.head())
display (bank_df.info())
display(bank_df.describe())
display (bank_df["y"].value_counts())

plt.figure(figsize=(10, 6))
sns.histplot(data=bank_df, x="age", bins= 10, hue = "y", multiple ="stack")
plt.title("Age distribution")

plt.figure(figsize=(10, 6))
yes_by_job = bank_df[bank_df["y"] == "yes"]["job"].value_counts()
sns.barplot(x=yes_by_job.index, y=yes_by_job.values)
plt.title ("Subscriptions by job")
plt.xticks(rotation=45)

plt.figure(figsize=(10, 6))
sns.boxplot(x="y", y="duration", data=bank_df)
plt.title("Call duration by subscription")
````

````python
#b Preprocessing
#to know exactly which preprocess use and how for each column
text_columns=["job", "marital", "education", "default", "housing", "loan", "contact", "month", "day_of_week","poutcome", "y"]

for column in text_columns:
    unique_values = bank_df[column].unique()
    print(f"Colonne '{column}' : {unique_values}")
````

````python
print("Type de bank_df['education'] :", type(bank_df["education"]))
print("Type de bank_df[['education']] :", type(bank_df[["education"]]))
print("Valeurs uniques :", bank_df["education"].unique())
````

````python
#b-preprocessing

from sklearn.preprocessing import OneHotEncoder, LabelEncoder, OrdinalEncoder, MinMaxScaler

label_encoder=LabelEncoder() #don't take the "unknown" value
bank_df["contact"]=label_encoder.fit_transform(bank_df["contact"])
bank_df["poutcome"]=label_encoder.fit_transform(bank_df["poutcome"])

ordinal_encoder_edu=OrdinalEncoder(categories=[["unknown", "illiterate", "basic.4y",
                                                "basic.6y", "basic.9y", "high.school",
                                                "professional.course", "university.degree"]])
bank_df["education"]=ordinal_encoder_edu.fit_transform(bank_df[["education"]])

ordinal_encoder_month=OrdinalEncoder(categories=[["mar", "apr", "may", "jun","jul", "aug",
                                                  "sep",  "oct", "nov", "dec"]])
bank_df["month"]=ordinal_encoder_month.fit_transform(bank_df[["month"]])[:, 0]

ordinal_encoder_day=OrdinalEncoder(categories=[["mon", "tue", "wed", "thu", "fri"]])
bank_df["day_of_week"]=ordinal_encoder_day.fit_transform(bank_df[["day_of_week"]])

bank_df["job"]=bank_df["job"].map({"blue-collar":0, "services":1, "admin.":2,
                                   "entrepreneur":3, "self-employed":4, "technician":5,
                                   "management":6, "student":7, "retired":8,
                                   "housemaid":9, "unemployed":10, "unknown":-1})
bank_df["default"]=bank_df["default"].map({"yes":1,"no":0, "unknown":-1})
bank_df["housing"]=bank_df["housing"].map({"yes":1,"no":0, "unknown":-1})
bank_df["marital"]=bank_df["marital"].map({"married":2, "single":1, "divorced":3, "unknown":0})
bank_df["loan"]=bank_df["loan"].map({"yes":1,"no":0, "unknown":-1})
bank_df["y"]=bank_df["y"].map({"yes":1,"no":0})

bank_df.head()
````
